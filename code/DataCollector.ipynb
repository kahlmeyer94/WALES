```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pickle
import os

def collect_paths(n_paths, store_path, refresh_rate = 60):
        # initial dataset
        if os.path.exists(store_path):
            with open(store_path, 'rb') as handle:
                ds = pickle.load(handle)
        else:
            ds = {}
            
        old_nmb_paths = len(ds)    
        
        # collect data
        browser = webdriver.Firefox()
        browser.get(('https://www.thewikigame.com/'))
        browser.find_element_by_css_selector('.btn.btn-primary').click()
        
        n_seconds = 1
        while len(ds)<n_paths:
            n_seconds = n_seconds%refresh_rate
            if n_seconds == 0:
                browser.refresh()
            res = browser.find_elements_by_class_name('link.round-challenge-font')
            if len(res)>0:
                start = res[0].text.lower()
                target = res[1].text.lower()
                if not (start, target) in ds:
                    
                    # scrape!
                    paths = []
                    searching = True
                    count = 0
                    while searching:
                        path = []
                        l = browser.find_elements_by_class_name(f'ng-tns-c11-{count}.ng-star-inserted')
                        for idx, element in enumerate(l):
                            if idx == len(l)-1:
                                s = element.text
                                path.append(s.lower())
                            elif idx%2==0:
                                s = element.text[:-2]
                                path.append(s.lower())
                            
                        searching = len(l)>0
                        count += 1
                        if searching:
                            paths.append(path)
                    if len(paths)>0:
                        ds[(start, target)] = paths
                        # save dataset
                        with open(store_path, 'wb') as handle:
                            pickle.dump(ds, handle)
                        print(f'Added {len(paths)} paths from {start} to {target}\nNumber of challenges: {len(ds)}')
            

            time.sleep(1)
            n_seconds += 1
        browser.close()
        return ds
    
store_path = 'human_results.p'
ds = collect_paths(5000, store_path, refresh_rate=60)
```
